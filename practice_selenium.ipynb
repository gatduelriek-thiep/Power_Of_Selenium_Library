## Selenium scraping

The purpose of this Python script is to scrape data from a website and export them into an Excel file

```python
# Gatduel use selenium library and import it to scrape data
from selenium import webdriver as wd
```

```python
from selenium.webdriver.common.by import By as b
```

## Open browser using the below process

```python
driver = wd.Chrome()
```

```python
# Get URL of the website you want to scrape
url = "https://www.scrapethissite.com/pages/simple/"
```

```python
driver.get(url)
```

```python
# Get the country list using Selenium b.CLASS_NAME
country_list = driver.find_elements(b.CLASS_NAME, "country-name")
# country_list
```

```python
# iterate through the country_list variable
country = []
for countries in country_list:
    temp = countries.text
    country.append(temp)

# country
```

```python
# Get population list using the above process
population_list = driver.find_elements(b.CLASS_NAME, "country-population")
```

```python
# Iterate through the process again
populations = []
for population in population_list:
    temp = population.text
    populations.append(temp)

# populations
```

```python
# This is the same process as show in the above 2 step
country_capital = driver.find_elements(b.CLASS_NAME, "country-capital")
```

```python
# Relate it to the above steps
capital = []
for capitals in country_capital:
    temp = capitals.text
    capital.append(temp)

# capital
```

```python
Area = driver.find_elements(b.CLASS_NAME, "country-area")
```

```python
# Here we are doing the same things
areas = []
for country_area in Area:
    temp = country_area.text
    areas.append(temp)
```

```python
# areas
```

```python
# Gatduel import pandas as p to give the scrape data structure and export them in to excel file
import pandas as p
```

```python
df = p.DataFrame()
df["Country Names"] = country
df["Populations"] = populations
df["Area"] = areas
df["Capital Cities"] = capital
```

```python
# df.head(8)
```

## Export data to an Excel file with the name selenium

```python
df.to_excel("selenium.xlsx", index = False)
```
